```python
import theano.tensor as T 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
dataset=pd.read_csv("Churn_Modelling.csv") 
```

```python
dataset.head() 
```

```python
x = dataset[['CreditScore', 'Age', 'Tenure','Balance',"NumOfProducts","HasCrCard","IsActiveMember","EstimatedSalary"]]
y = dataset[['Exited']] 
```

```python
x.head(10)  
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer 
```

```python
preprocess = make_column_transformer(
    (['CreditScore', 'Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary'], StandardScaler()),) 
```

```python
preprocess.fit_transform(x)[:10] 
```

```python
from keras.models import Sequential 
from keras.layers import Dense
from keras.layers import Dropout
from keras.regularizers import l2 
```

```python
from sklearn.model_selection import StratifiedKFold 
import numpy
seed = 7
numpy.random.seed(seed) 
```

```python
kfold = StratifiedKFold(n_splits = 10, shuffle = True, random_state=seed)
cvscores = [] 
```

```python
for train, test in kfold.split(x,y):
    classifier = Sequential()
    classifier.add(Dense(6, input_dim=8, activation = "relu", kernel_regularizer=l2(0.001)))
    classifier.add(Dropout(0.2))
    classifier.add(Dense(6, activation = "relu"))
    classifier.add(Dense(1, activation = "sigmoid"))
    classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
    classifier.fit(x.loc[train], y.loc[train], batch_size = 10, nb_epoch = 100)  
    scores = classifier.evaluate(x.loc[test], y.loc[test], verbose=0)
    print("%s: %.2f%%" % (classifier.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)
print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores))) 
```
